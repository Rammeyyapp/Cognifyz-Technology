# Task 5

```python
reasons_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']
combined_reasons = data[reasons_columns].apply(lambda row: ' '.join(row), axis=1)
```

```python
from collections import Counter
import re 
def tokenize_reasons(reasons_list):
    reasons_text = ' '.join(reasons_list).lower()
    tokens = re.findall(r'\b\w+\b', reasons_text)
    return Counter(tokens)

reasons_summary = tokenize_reasons(combined_reasons)

top_reasons = reasons_summary.most_common(15)

print("Top 10 reasons for investment choices:")
for reason, count in top_reasons:
    print(f" - {reason}: {count}")
```

# Task 8

```python
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
data['Duration_Numeric'] = data['Duration'].map(duration_mapping)
average_duration = data['Duration_Numeric'].mean()
print(f"Average Investment Duration: {average_duration:.2f} years")
```
